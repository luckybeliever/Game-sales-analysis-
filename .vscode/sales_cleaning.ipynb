So there is no null value in these columns NA_Sales	,EU_Sales ,JP_Sales	,Other_Sales ,Global_Sales

```python
df.notna()
```

```python
df_nonull = df[df.notna().all(axis=1)]
df_nonull
```

This is table which have zero null values in it

```python
df_nonull.drop_duplicates(inplace=True)
```

```python
df_nonull.columns
```

```python
df_nonull.index.name = 'Game_sale_index'
df_nonull.drop(columns='Rank', axis = 0,inplace=True)
df_nonull
```

```python
df_nonull.reset_index(drop='Game_sale_index',inplace= True)
df_nonull.index.name = 'Game_sale_index'
df_nonull
```

Here the values of sales in million dollars


```python
type(df_nonull.Year)
```

```python
df_nonull['Year'].dtype
```

```python
df_nonull['Year'] = df_nonull['Year'].astype(int)
df_nonull.info()
```

```python
df_nonull
```

So here data cleaning is done 


```python
df_cleaned = df_nonull.copy()
```

```python
df_cleaned.to_csv("Game_sales_cleaned.csv", index=True)
```
